```python
#class object
class Student: #defining class
  def __init__(self,name,age): #constructor runs when new student is created
    self.name = name #stores the name in obj
    self.age = age   #stores the age in obj
  def introduce(self):  #method to print student info
    print(f"my name is {self.name}, i am {self.age} years old.")
#creating objects
s1 = Student("aastha",21)
s1.introduce()
```

```python
#inheritance
class shape(): #define class
  def area(self): #define method area
    print("calculating area...")
class square(shape): #define class square inherited from shape
  def __init__(self,side): #constructor for creating obj
    self.side=side
  def area(self):
    print("area of square" , self.side*self.side)
s = square(2)
s.area()
```

```python
class shape():
  def area(self):
    print("calculating area")
class square(shape):
  def __init__(self,side):
    self.side=side
  def area(self):
    print("area of square ", self.side*self.side)
s=square(4)
s.area()
```

```python
#polymorphism
#compile time using default arguments
class calculator:
  def add(self,a,b=0,c=0):
    return a+b+c
calc=calculator()
print(calc.add(8))
print(calc.add(2,5))
```

```python
# using *args
class calculator:
  def add(self,*numbers):
    return sum(numbers)
calc=calculator()
print(calc.add(4,4))
```

```python
#run time polymorphism
class animal:
  def sound(self):
    print("animal makes sound")
class dog(animal):
  def sound(self):
    print("dog barks")
class cat(animal):
  def sound(self):
    print("cat meows")
def make_sound(pet):
  pet.sound()
make_sound(animal())
make_sound(cat())
make_sound(dog())
```

```python
#encapsulation
class account:
  def __init__(self):
    self.__balance = 1000 #private variable
  def deposit(self,amount):
    if amount > 0:
      self.__balance += amount
  def get_balance(self):
    return self.__balance
acc = account()
acc.deposit(1000)
print(acc.get_balance())
```

```python
#abstration
from abc import ABC, abstractmethod #This line is used in Python to create abstract classes and abstract methods.
#(@abstractmethod) A decorator used to declare a method that must be implemented in any child class.
class ATM(ABC):
  @abstractmethod
  def deposit(self,amount):
    pass
  @abstractmethod
  def withdraw(self,amount):
    pass
  @abstractmethod
  def check_balance(self):
    pass
class MyATM(ATM):    # 'MyATM' inherits from 'ATM' and must implement all abstract methods
  def __init__(self): # Constructor to initialize object
    self.balance = 0 # Initialize account balance to 0

  def deposit(self,amount): # Implement deposit method
    self.balance += amount  # Add amount to balance
    print(f"₹{amount} deposited successfully.") # Show success message


  def withdraw(self,amount): # Implement withdraw method
    if amount <= self.balance:  # Check if sufficient balance
      self.balance -= amount   # Subtract amount from balance
      print(f"₹{amount} withdrawn successfully.")  # Show success message
    else:
      print("insufficient amount") # Show error if not enough balance
  def check_balance(self):   # Implement check_balance method
    print(f"available balance: ₹{self.balance}")  # Display current balance

atm = MyATM()  # Create an instance of MyATM

atm.deposit(1000)
atm.check_balance()
atm.withdraw(500)
atm.check_balance()
atm.withdraw(700)
```
